### (5) 특징 추출(feature extraction)

- 그룹화 및 정렬 기준에 맞게 추출할 수 있다.

```python
# 05
# (5) 특징 추출(feature extraction)
from tsfresh import extract_features

extracted_festures = extract_features(timeseries, column_id= "id", column_sort= 'time')
```

```python
# 1. 결측치 및 무한값 획인 - 히트맵(결측치 시각화)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.heatmap(extracted_festures.isna(), cbar=False, cmap="viridis")

plt.title("결측치(NaN) 히트맵")
plt.show()
```

### (6) impute
- 특징 추출된 모든 값을 동일한 열의 중앙/극단값으로 바꾼다
  - -inf -> min
  - +inf -> max
  - NaN -> median

### (7)  select_features(X, y)
-  특징 행렬 X의 모든 특징(열)의 중료성을 확인하고,
- 관련 특징만 포함하는 측소된 특징 행렬을 반환(필터링)

```python

from tsfresh import select_features
from tsfresh.utilities.dataframe_functions  import impute

# (6) impute
impute(extracted_festures)
```

```python
# (7) select_features(X, y)
features_filtered = select_features(extracted_festures, y)
```
