```python
# umpy 호출 및 Numpy array 생성: np.array([...])

import numpy as np
import pandas as pd

a =  np.array([1, 2,3])
print(a)
```

```python
# Concatenate를 통해 Numpy array 합치기: np.concatenate((a,b))
a = np.array([1, 2,3, 4])
b = np.array([5, 6, 7, 8])
print(np.concatenate((a, b)))
```

```python
# 슬라이싱과 인덱싱: array[a:b]
data = np.array([1, 2, 3])

print(data[1])
print(data[0:2])
print(data[1:])
print(data[-2:])
```

```python
# Numpy 연산: np.max(), np.min(), np.sum()

print(data.max())
print(data.min())
print(data.sum())

```

```python
# iloc와 loc
import pandas as pd
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Kim', 'Park', 'Hong' ],
    'Age': [25, 30, 35, 40, 45, 50]
}, index=[101, 102, 103, 104, 105, 106])

print(df)

print("========")
print(df.loc[102, 'Age']  )       # 인덱스가 102인 행의 'Age' 컬럼 값
print(df.iloc[1, 1] )             # (두 번째 행, 두 번째 열)

```

```python
dates = pd.date_range("20130101", periods = 6)      # periods는 20130101을 포함하여 총 6개의 날짜(년월일) index를 생성해줍니다.
print(dates)
```
