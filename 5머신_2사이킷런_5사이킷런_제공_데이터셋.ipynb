### 데이터셋 선정

- 회귀하기 위한 데이터셋 : load_diabetes()

```python
#  당뇨병 데이커 불러오기
from sklearn.datasets  import  load_diabetes
dataset = load_diabetes()
```

```python
# 데이터프레임으로 만들기
cancer_df = pd.DataFrame(data=dataset.data, columns=dataset.feature_names)
cancer_df
```
