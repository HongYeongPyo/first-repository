```python
# 시계열 데이터의 변동성 군집(Volatility Clustering) 개념을 실험적으로 생성하고,
# 이에 대해 자기상관 함수(ACF, Autocorrelation Function)를 계산하여 확인하는 실습

import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf               # acf: 시계열 데이터의 자기상관 함수(AutoCorrelation Function)를 계산하는 함수


# 시계열 데이터 생성
# 랜덤 시드 고정                                        # 랜덤성을 제어하여 결과 재현성(reproducibility) 확보
np.random.seed(42)                                      # 특정한 lag=2에서 강한 상관성이 나타나도록 설계


# 수익률 시계열 생성: 평균 0, 분산 1의 정규분포         # 평균 0, 표준편차 1인 정규분포 난수 300개 생성. 시계열적인 수익률 데이터로 간주
n = 300
r = np.random.normal(0, 1, n)


# 변동성 클러스터링을 만들어보자: 중간에만 분산 증가     # 금융 시계열에서 자주 나타나는 "변동성 군집" 현상을 인위적으로 생성즉, 일정 구간에만 높은 변동성을 가지는 시계열 구성
r[100:200] *= 3  # 중간 구간의 변동성만 크게 만듦        # 시계열의 **중간 구간(100~199)**만 변동성을 3배로 증폭


# 수익률의 제곱                                          # 보통 변동성은 수익률의 절댓값이나 제곱값으로 측정됨. 이를 통해 평균은 0에 가깝지만, 변동 폭이 큰 구간을 강조함
r_squared = abs(r)

# ACF 계산
acf_r = acf(r, nlags=20)                                 # acf_r: 원래 수익률 시계열의 자기상관 계산 → 이론적으로는 백색잡음에 가까워야 함
acf_r_squared = acf(r_squared, nlags=20)                 # 절댓값 시계열의 ACF → 변동성 클러스터링이 있을 경우 양(+)의 자기상관이 나타남




#=============================
# 평균 0, 분산 1의 정규분포에서 생성한 수익률 시계열에 대해 중간 구간의 분산을 인위적으로 증가시켜 변동성 클러스터링을 시뮬레이션하였다
# 이후 자기상관 함수(ACF)를 계산한 결과, 원래의 수익률 시계열은 자기상관이 거의 없으나,
# 절댓값 수익률 시계열에서는 유의미한 양의 자기상관이 나타나, 변동성 군집의 존재를 통계적으로 시사
```

```python
# 03

# 시각화
fig, axes = plt.subplots(3, 1, figsize=(12, 10), sharex=False)

# 수익률 시계열
axes[0].plot(r, color='blue')
axes[0].set_title("Return r (volatility clusters)")                             # 수익률 r (변동성 클러스터 있음
axes[0].axvspan(100, 200, color='orange', alpha=0.2, label="Volatility increasing range")   # 변동성 증가 구간
axes[0].legend()


# 수익률 시계열의 제곱
axes[1].plot(r_squared, color='red')
axes[1].set_title("Return squared r²")                                             # 수익률 제곱


# ACF 비교
axes[2].stem(acf_r, linefmt='b-', markerfmt='bo', basefmt=' ', label='ACF of r')
axes[2].stem(acf_r_squared, linefmt='r-', markerfmt='ro', basefmt=' ', label='ACF of r²')  # ACF: 수익률 vs 수익률 제곱
axes[2].set_title("ACF: Return vs Return squared")
axes[2].legend()

plt.tight_layout()
plt.show()
```

```python
# 05
# 시계열 데이터 생성 → 자기상관 함수(ACF) 시각화 과정으로, lag=2에서 인위적으로 강한 상관성을 갖도록 설계된 시계열

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf


# 시계열 데이터 생성. 특정한 lag=2에서 강한 상관성이 나타나도록 설계
np.random.seed(42)
n = 100                                          # 길이 100짜리 시계열
base_noise = np.random.normal(0, 1, n)           # 평균 0, 표준편차 1의 정규분포 노이즈


# 현재값은 2칸 전 값에 의존 + 노이즈. AR(2) 모델의 한 변형
series = np.zeros(n)
for i in range(2, n):                                # lag=2에서 매우 강한 자기상관성이 있도록 설계됨
    series[i] = 0.9 * series[i-2] + base_noise[i]    # 시계열 생성 규칙: 현재 값은 2시점 전 값의 0.9배 + 노이즈


# 시리즈를 데이터프레임으로 포장
df = pd.Series(series)                              # plot_acf()는 pandas.Series 타입을 입력으로 받는 것이 일반적


# ACF 시각화
plt.figure(figsize=(10, 4))
plot_acf(df, lags=20)                              # plot_acf(df, lags=20): 최대 20 시차까지의 자기상관 시각화

plt.title("ACF plot (Strong correlation at lag=2)")
plt.tight_layout()
plt.show()


#===============================
#  ??  lag=2에서 높은 막대(자기상관 계수)가 나타날 것이며, 이는 시계열 생성 구조에 부합함
```

```python
#  06

from statsmodels.graphics.tsaplots import plot_pacf

# PACF 시각화
plt.figure(figsize=(10, 4))
plot_pacf(df, lags=20, method='ywm')

plt.title("PACF plot (Strong partial autocorrelation at lag=2)")
plt.tight_layout()
plt.show()
```

```python
# 16

# ⑥ ARIMA 적용
model1 = ARIMA(data, order=(1,0,0))        # AR(1) - 현재 값이 직전 값에만 의존	자기상관 있음, 오차는 백색잡음
model2 = ARIMA(data, order=(0,0,1))        # MA(1) - 현재 값이 직전 오차항에만 의존	충격(오차)이 다음 시점에 영향
model3 = ARIMA(data, order=(1,0,1))        # ARMA(1,1) - AR과 MA 요소 모두 포함	자기회귀 + 오차의 영향 결

res1 = model1.fit()                        # 각 .fit() 결과는 모델의 추정 결과 객체이며, 다음을 포함
res2 = model2.fit()                        # - 회귀계수 (.params) 로그우도, AIC/BIC 등 모델 평가 지표 (.aic, .bic)
res3 = model3.fit()                        # - 예측 결과 (.forecast(), .predict())


#===================
# order(p, d, q)
# p	AR (자기회귀, AutoRegressive)	- 과거 시계열 값에 대한 의존도
# d	차분 차수 (Integration)	 - 비정상성을 제거하기 위한 차분 횟수
# q	MA (이동평균, Moving Average)	- 과거 오차에 대한 의존도

# d=0으로 모두 설정한 이유: 데이터가 정상성(stationary)을 만족한다고 가정한 경우. 비정상 시계열일 경우 d>0이 필요


# ===========================
# 시계열 데이터의 모형화를 위해 ARIMA 모형의 세 가지 구조(AR, MA, ARMA)를 비교 적용하였다
# 각각의 모델은 시계열 데이터의 자기회귀성, 오차 구조 또는 그 결합을 설명하는 데 유용하며,
# 학습자 행동 예측 및 패턴 탐지에 적합한 구조를 탐색하기 위한 기초 분석으로 활용
```

```python
# 17

# ⑦ ARIMA 결과 확인
print(res1.summary())


#=======================================
# 모델의 적합도, 계수의 유의성, 잔차 진단 등을 포함

# 1.  기본 정보
# Model: ARIMA(1,0,0)	자기회귀(AR) 1차 모형, 차분(d)=0, 이동평균 없음
# No. Observations: 60	총 60개의 시계열 관측값
# Log Likelihood: -354.181	로그우도 값 (높을수록 모델 적합도 좋음)
# AIC / BIC / HQIC	정보 기준(Akaike/Bayesian/Hannan-Quinn) – 작을수록 우수한 모델


# 2. 회귀 계수 해석 (coef, P>|z|)
# const | 301.04 | 0.000 | 유의미한 상수항 (평균 수준) |
# | ar.L1 | 0.0804 | 0.667 | 유의하지 않음, 시차 1에 의존성 없음 |
# | sigma2 | 7848.57 | 0.000 | 오차항의 분산 (변동성) |

# P>|z| < 0.05이면 통계적으로 유의
#여기서는 ar.L1의 p-value가 0.667 > 0.05로 유의하지 않음 → AR(1) 모형의 필요성이 낮음


# 3. 잔차 진단
# Ljung-Box (Q): Prob(Q)=0.99	- 잔차에 자기상관 없음 → 좋은 징후
# Heteroskedasticity (H): Prob=0.18	- 이분산성 없음 (가정 만족)
#Jarque-Bera (JB): Prob=0.00	- 잔차의 정규성 불만족 (왜도, 첨도 크기 때문)
# → 잔차가 비정규분포일 가능성 존재하나, 자기상관·이분산 문제는 없음


# ====
# **AR(1) 계수(ar.L1)**가 통계적으로 유의하지 않음 → 시차 1의 자기회귀 구조 필요성이 낮음
# - AIC/BIC 값은 다른 모델과 비교할 기준이 됨
# - 잔차는 자기상관과 이분산은 없지만 정규성은 불충분
# - 모델 성능 자체는 안정적이지만, AR 요소가 크게 기여하지 않는 것으로 해석


# =======
# ARIMA(1,0,0) 모형을 적용한 결과, 상수항은 유의하게 나타났으나
# - 시차 1의 자기회귀 계수는 유의하지 않아 자기회귀 구조의 설명력이 낮은 것으로 나타났다(p=0.667)

# AIC=714.362로, 다른 모형과의 비교를 통해 적절성을 평가할 수 있으며,
# - 잔차에 대한 Ljung-Box 및 Heteroskedasticity 테스트 결과는 모델의 기본 가정을 만족하는 것으로 나타났다.
```

```python
# 20 - 위의 3가지 모델 분석 비교 표

# 세 가지 ARIMA 모델 ARIMA(1,0,0), ARIMA(0,0,1), ARIMA(1,0,1)의 적합 결과를
# AIC, BIC, Log Likelihood, 유의확률(P-value) 등의 지표로 한눈에 비교 출력하는 코드

import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import pprint
from statsmodels.tsa.stattools import adfuller # 필요하다면 추가

# 수정된 extract_summary 함수
def extract_summary(model, name):
    try:
        # test_serial_correlation 결과 가져오기
        ljungbox_test = model.test_serial_correlation(method='ljungbox', lags=[1])
        # 반환 타입 확인 및 값 추출 방식 변경
        if isinstance(ljungbox_test, pd.DataFrame):
             ljungbox_p = ljungbox_test.iloc[0]['lb_pvalue']
        elif isinstance(ljungbox_test, tuple) and len(ljungbox_test) > 1:
             # 튜플로 반환되는 경우 (e.g., (statistic, pvalue, ...) )
             ljungbox_p = ljungbox_test[1][0] # p-value 값 추출 (인덱스는 반환 값 구조에 따라 다를 수 있음)
        else:
             ljungbox_p = None # 또는 다른 기본값

        # test_normality 결과 가져오기
        normality_test = model.test_normality(method='jarquebera')
        # 반환 타입 확인 및 값 추출 방식 변경
        if isinstance(normality_test, pd.DataFrame):
             jb_p = normality_test.iloc[0]['jb_pvalue']
             skew = normality_test.iloc[0]['skew']
        elif isinstance(normality_test, tuple) and len(normality_test) > 1:
             # 튜플로 반환되는 경우 (e.g., (jb_statistic, jb_pvalue, skew, kurtosis, ...) )
             jb_p = normality_test[1]  # p-value 값 추출 (인덱스는 반환 값 구조에 따라 다를 수 있음)
             skew = normality_test[2] # skew 값 추출 (인덱스는 반환 값 구조에 따라 다를 수 있음)
        else:
             jb_p = None
             skew = None

    except Exception as e:
        print(f"Error processing model {name}: {e}")
        ljungbox_p = None
        jb_p = None
        skew = None


    return {
        'Model': name,
        'Log Likelihood': model.llf,
        'AIC': model.aic,
        'BIC': model.bic,
        'AR(1) pval': model.pvalues.get('ar.L1', None),
        'MA(1) pval': model.pvalues.get('ma.L1', None),
        'sigma2': model.params.get('sigma2', None),
        'LjungBox p': ljungbox_p,
        'JB p': jb_p,
        'Skew': skew
    }

# 모델 적합 (기존 코드와 동일)
# ... (이전 셀에서 model1, model2, model3 정의 및 fit() 호출)

# 요약 함수 사용
results = pd.DataFrame([
    extract_summary(model1, 'ARIMA(1,0,0)'),
    extract_summary(model2, 'ARIMA(0,0,1)'),
    extract_summary(model3, 'ARIMA(1,0,1)')
])

# AIC 기준 정렬
results = results.sort_values('AIC').reset_index(drop=True)

# 보기 좋게 출력
pprint.pprint(results)


#====================
# 1. 모델 적합도 비교 (Log Likelihood, AIC, BIC)
# AIC와 BIC 모두 ARIMA(0,0,1)이 가장 낮음 → 가장 적합한 모델로 평가됨.
# - ARIMA(1,0,1)은 오히려 과적합(overfitting) 가능성 있음 (AIC, BIC 모두 가장 큼)

#  2. 회귀 계수 유의성 (p-value)
# 모든 계수의 p-value > 0.05 → 통계적으로 유의하지 않음.
# - 이는 데이터에 명확한 자기상관(AR)이나 오차구조(MA)가 없을 수 있음을 시사

# 3. 오차 분산 (sigma2)
# 오차 분산은 **ARIMA(1,0,1)**이 가장 작지만, AIC는 높아 전체 모델 성능은 상대적으로 낮음.
# - 오차분산 차이는 크지 않음 → 모델 선택에 큰 영향 X

# 4. Ljung-Box, JB Test, Skew/Kurtosis
# 현재 테이블에선 'None'으로 표기되어 있어 잔차 진단 결과가 누락됨.
# - 이 부분을 보완하면 잔차의 정상성, 정규성, 자기상관 여부도 함께 판단 가능해집니다.


# === 종합 분석 --
# AIC/BIC 기준 적합도	-  ARIMA(0,0,1) 최적
# 계수 유의성	❌ 모든 모델에서 비유의 (p > 0.05)
# 모델 단순성	-  ARIMA(0,0,1) 단순 + 성능 우수
# 잔차 진단	-  현재 정보 부족 (None)

# 세 가지 ARIMA 모델을 비교한 결과, AIC 및 BIC 기준으로는 ARIMA(0,0,1) 모델이 가장 적합한 것으로 나타났다.
# - 이는 데이터가 과거 오차에 기반한 단기적인 영향만을 반영하며, 자기회귀 구조는 두드러지지 않음을 의미한다
# - 다만 모든 모델에서 계수의 유의성이 확보되지 않아 예측 모형으로의 활용에는 주의가 필요하며, 향후 잔차 진단을 통해 보다 정밀한 평가가 요구된다.
```
